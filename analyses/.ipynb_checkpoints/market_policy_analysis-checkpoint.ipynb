```python
def convert(yearly_bau_results_df):
    """ Filtering GHG emissions from BAU """

    data_GHG_household = yearly_bau_results_df['GHG Emissions (tCO2-e) per household']

    """ Converts household data in tCO2e into total residential data in MtCO2e"""
    data_GHG_residential = (data_GHG_household * yearly_bau_results_df['Number of Domestic Consumers (households)']) / 1000000.0

    """ Applies scaling factor used in simulation, so that data represents total emissions, not only residential (30%)"""
    data_GHG_total = data_GHG_residential / 0.3

    return pd.DataFrame(data_GHG_total)

GHG_total_df = convert(yearly)
GHG_total_df=GHG_total_df.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e)'})[:]
GHG_total_df3 = convert(yearly3)
GHG_total_df3=GHG_total_df3.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 3$ Carbon Pricing'})[:]
GHG_total_df8 = convert(yearly8)
GHG_total_df8=GHG_total_df8.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 8$ Carbon Pricing'})[:]
GHG_total_df13 = convert(yearly13)
GHG_total_df13=GHG_total_df13.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 13$ Carbon Pricing'})[:]
GHG_total_df18 = convert(yearly18)
GHG_total_df18=GHG_total_df18.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 18$ Carbon Pricing'})[:]
GHG_total_df23 = convert(yearly23)
GHG_total_df23=GHG_total_df23.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 23$ Carbon Pricing'})[:]
GHG_total_df28 = convert(yearly28)
GHG_total_df28=GHG_total_df28.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 28$ Carbon Pricing'})[:]
```

## Tariffs Validation

1. load historic and simulated yearly average tariffs
2. plot historic and simulated results
3. use a descriptive statistic measure to validate results from simulation against historic values

Different sources where used to construct a historic dataset of average electricity tariffs.
- Historic data found in the [ACCC report](https://www.accc.gov.au/system/files/Victorian%20Electricity%20Distribution%20Networks.pdf) from 2017. This data is from a household in VIC that use 4000 KWh/year, has gas water heating. The report has data for the years 1995, 2001, 2002, and from 2007 to 2017. But data is used from 2001. Since the data is given as total bill per year in /$, the price per KWh is computed as (AnnualBill/4000KWh)*100, to get the value in c/kWh.
- [St. Vinnies historic data](https://www.vinnies.org.au/page/Our_Impact/Incomes_Support_Cost_of_Living/Energy/VIC/) on tariffs per retailer and distribution area. This data has been collected since 2010, every quarter of the year. I used the ‘first block’ charges to calculate the average charge of usage in c/KWh. 

Prices from the reports and data sources are adjusted to 2019 values as simulatied tariffs are presented in real terms for the base year of 2019.

```python
def convert(yearly_bau_results_df):
    """ Filtering GHG emissions from BAU """

    data_GHG_household = yearly_bau_results_df['GHG Emissions (tCO2-e) per household']

    """ Converts household data in tCO2e into total residential data in MtCO2e"""
    data_GHG_residential = (data_GHG_household * yearly_bau_results_df['Number of Domestic Consumers (households)']) / 1000000.0

    """ Applies scaling factor used in simulation, so that data represents total emissions, not only residential (30%)"""
    data_GHG_total = data_GHG_residential / 0.3

    return pd.DataFrame(data_GHG_total)

GHG_total_df = convert(yearly)
GHG_total_df=GHG_total_df.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e)'})[:]
GHG_total_df30 = convert(yearly30)
GHG_total_df30=GHG_total_df30.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 30% renewable'})[:]
GHG_total_df50 = convert(yearly50)
GHG_total_df50=GHG_total_df50.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 50% renewable'})[:]
GHG_total_df70 = convert(yearly70)
GHG_total_df70=GHG_total_df70.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 70% renewable'})[:]
GHG_total_df90 = convert(yearly90)
GHG_total_df90=GHG_total_df90.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with 90% renewable'})[:]
```

```python
def convert(yearly_bau_results_df):
    """ Filtering GHG emissions from BAU """

    data_GHG_household = yearly_bau_results_df['GHG Emissions (tCO2-e) per household']

    """ Converts household data in tCO2e into total residential data in MtCO2e"""
    data_GHG_residential = (data_GHG_household * yearly_bau_results_df['Number of Domestic Consumers (households)']) / 1000000.0

    """ Applies scaling factor used in simulation, so that data represents total emissions, not only residential (30%)"""
    data_GHG_total = data_GHG_residential / 0.3

    return pd.DataFrame(data_GHG_total)

GHG_total_df = convert(yearly)
GHG_total_df=GHG_total_df.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e)'})[:]
GHG_total_df0 = convert(yearly0)
GHG_total_df0=GHG_total_df0.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with Price Merit Order'})[:]
GHG_total_df23 = convert(yearly23)
GHG_total_df23=GHG_total_df23.rename(columns = {0:'Simulated_GHGE_VIC (MtCO2e) with Emission Merit Order'})[:]
```

Different sources where used to construct a historic dataset of average electricity tariffs.
- Historic data found in the [ACCC report](https://www.accc.gov.au/system/files/Victorian%20Electricity%20Distribution%20Networks.pdf) from 2017. This data is from a household in VIC that use 4000 KWh/year, has gas water heating. The report has data for the years 1995, 2001, 2002, and from 2007 to 2017. But data is used from 2001. Since the data is given as total bill per year in /$, the price per KWh is computed as (AnnualBill/4000KWh)*100, to get the value in c/kWh.
- [St. Vinnies historic data](https://www.vinnies.org.au/page/Our_Impact/Incomes_Support_Cost_of_Living/Energy/VIC/) on tariffs per retailer and distribution area. This data has been collected since 2010, every quarter of the year. I used the ‘first block’ charges to calculate the average charge of usage in c/KWh. 

Prices from the reports and data sources are adjusted to 2019 values as simulatied tariffs are presented in real terms for the base year of 2019.
