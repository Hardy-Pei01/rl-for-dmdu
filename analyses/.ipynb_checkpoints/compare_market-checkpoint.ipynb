## Load the evaluatation results for EA

```python
import os, sys
sys.path.append(os.path.abspath('../'))
import pandas as pd
import numpy as np
from EMAWorkbench import load_results

results_path = '../results/market_performance/'
files = os.listdir(results_path)
files
```

```python
def compute_error_bar(df):
    u_mean=np.mean(all_outcomes["price_reduction"])
    u_25=np.percentile(all_outcomes["price_reduction"], 25)
    u_75=np.percentile(all_outcomes["price_reduction"], 75)
    r_mean=np.mean(all_outcomes["emission_reduction"])
    r_25=np.percentile(all_outcomes["emission_reduction"], 25)
    r_75=np.percentile(all_outcomes["emission_reduction"], 75)
    return (u_mean, (u_mean-u_25, u_75-u_mean), r_mean, (r_mean-r_25, r_75-r_mean))
```

```python
ea = []
for i in range(0, 105):
    experiments, all_outcomes = load_results(results_path + files[i])
    all_outcomes = dict((k, all_outcomes[k]) for k in ('price_reduction', 'emission_reduction'))
    mean_outcome = compute_error_bar(all_outcomes)
    ea.append(mean_outcome)
    print(files[i])
ea = pd.DataFrame(ea)
ea.columns = ["price_reduction", "p_inter", "emission_reduction", "e_inter"]
ea["Algorithms"] = "Epsilon-NSGA-II"
ea
```

```python
free = []
for i in range(135, 145):
    experiments, all_outcomes = load_results(results_path + files[i])
    all_outcomes = dict((k, all_outcomes[k]) for k in ('price_reduction', 'emission_reduction'))
    mean_outcome = compute_error_bar(all_outcomes)
    free.append(mean_outcome)
    print(files[i])
free = pd.DataFrame(free)
free.columns = ["price_reduction", "p_inter", "emission_reduction", "e_inter"]
free["Algorithms"] = "Robust DQN"
free
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
palette = sns.color_palette()
palette2 = sns.color_palette("Paired")
data = pd.concat([ea, free, epopt, deterministic, no1, no2, no3, no, nominal])
plt.figure(figsize=(9,6), dpi= 100)
sns.scatterplot(data=data, x="price_reduction", y="emission_reduction", hue="Algorithms", style="Algorithms", s=100, 
                palette=[palette[1], palette[3], palette[5], palette[6], palette[7], palette[8], palette[9], palette2[0], palette2[1]])
plt.xlabel("price-reduction ($/MWh)")
plt.ylabel("emission-reduction (tCO2-e per household)")
plt.show()
```

```python
fig = plt.figure(figsize=(12,12), dpi=100)
gs = fig.add_gridspec(2, 2, hspace=0, wspace=0)
ax = gs.subplots(sharex='col', sharey='row')
_, _, bars = ax[0,0].errorbar(x=ea['price_reduction'], y=ea['emission_reduction'], xerr=zip(*ea['p_inter']), 
                              yerr=zip(*ea['e_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[0,0].errorbar(x=no1['price_reduction'], y=no1['emission_reduction'], xerr=zip(*no1['p_inter']), 
                               yerr=zip(*no1['e_inter']), fmt='o', ls='none', label ='No Action')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

_, _, bars = ax[0,1].errorbar(x=ea['price_reduction'], y=ea['emission_reduction'], xerr=zip(*ea['p_inter']), 
                              yerr=zip(*ea['e_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[0,1].errorbar(x=free['price_reduction'], y=free['emission_reduction'], xerr=zip(*free['p_inter']), 
                               yerr=zip(*free['e_inter']), fmt='o', ls='none', label ='Robust DQN')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

_, _, bars = ax[1,0].errorbar(x=ea['price_reduction'], y=ea['emission_reduction'], xerr=zip(*ea['p_inter']), 
                              yerr=zip(*ea['e_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[1,0].errorbar(x=epopt['price_reduction'], y=epopt['emission_reduction'], xerr=zip(*epopt['p_inter']), 
                               yerr=zip(*epopt['e_inter']), fmt='o', ls='none', label ='EPOpt')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

_, _, bars = ax[1,1].errorbar(x=ea['price_reduction'], y=ea['emission_reduction'], xerr=zip(*ea['p_inter']), 
                              yerr=zip(*ea['e_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[1,1].errorbar(x=deterministic['price_reduction'], y=deterministic['emission_reduction'], xerr=zip(*deterministic['p_inter']), 
                               yerr=zip(*deterministic['e_inter']), fmt='o', ls='none', label ='Deterministic DQN')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

ax[0,0].set_title("(a)")
ax[0,1].set_title("(b)")
ax[1,0].set_title("(c)")
ax[1,1].set_title("(d)")
ax[0,0].set(ylabel='emission-reduction (tCO2-e per household)')
ax[1,0].set(xlabel='price-reduction ($/MWh)', ylabel='emission-reduction (tCO2-e per household)')
ax[1,1].set(xlabel='price-reduction ($/MWh)')
ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
ax[1,1].legend()
plt.show()
```
