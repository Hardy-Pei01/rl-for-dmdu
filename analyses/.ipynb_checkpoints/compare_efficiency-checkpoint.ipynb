# Cartpole Single

```python
import os, sys
sys.path.append(os.path.abspath('../'))
import pandas as pd

import numpy as np
import seaborn as sns
import pandas as pd

results_path = '../logs/cartpole_single/'
files = os.listdir(results_path)
```

```python
ea = pd.read_csv(results_path + files[5])[["epsilon_progress", "nfe"]]
ea["time_total_s"] = [i*0.183 for i in ea["nfe"]]
ea1 = pd.read_csv(results_path + files[6])[["epsilon_progress", "nfe"]]
ea1["time_total_s"] = [i*0.183 for i in ea1["nfe"]]
ea2 = pd.read_csv(results_path + files[7])[["epsilon_progress", "nfe"]]
ea2["time_total_s"] = [i*0.1836 for i in ea2["nfe"]]
ea3 = pd.read_csv(results_path + files[8])[["epsilon_progress", "nfe"]]
ea3["time_total_s"] = [i*0.1836 for i in ea3["nfe"]]
ea4 = pd.read_csv(results_path + files[9])[["epsilon_progress", "nfe"]]
ea4["time_total_s"] = [i*0.1842 for i in ea4["nfe"]]
ea = pd.concat([ea, ea1, ea2, ea3, ea4])
ea["time_total_s"] = [round(i / 500.0) * 500.0 for i in ea["time_total_s"]]
print(files[5])
print(files[6])
print(files[7])
print(files[8])
print(files[9])
ea
```

```python
deterministic = pd.read_csv(results_path + files[0] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
deterministic1 = pd.read_csv(results_path + files[1] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
deterministic2 = pd.read_csv(results_path + files[2] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
deterministic3 = pd.read_csv(results_path + files[3] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
deterministic4 = pd.read_csv(results_path + files[4] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
deterministic = pd.concat([deterministic, deterministic1, deterministic2, deterministic3, deterministic4])
deterministic["time_total_s"] = [round(i / 500.0) * 500.0 for i in deterministic["time_total_s"]]
print(files[0])
print(files[1])
print(files[2])
print(files[3])
print(files[4])
deterministic
```

```python
robust = pd.read_csv(results_path + files[15] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
robust1 = pd.read_csv(results_path + files[16] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
robust2 = pd.read_csv(results_path + files[17] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
robust3 = pd.read_csv(results_path + files[18] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
robust4 = pd.read_csv(results_path + files[19] + "/progress.csv")[["episode_reward_mean", "time_total_s"]]
robust = pd.concat([robust, robust1, robust2, robust3, robust4])
robust["time_total_s"] = [round(i / 500.0) * 500.0 for i in robust["time_total_s"]]
print(files[15])
print(files[16])
print(files[17])
print(files[18])
print(files[19])
robust
```

```python
def read_epopt(path):
    lines = []
    with open(path, "r") as f:
        for each_line in f:
            line = each_line.strip().split(" ")
            output_line = [float(line[0]), float(line[3]), float(line[5])]
            lines.append(output_line)
    df = pd.DataFrame(lines, columns=["average","10p", "time_total_s"])
    return df
epopt = read_epopt(results_path + files[10] + "/train_curve.txt")
epopt1 = read_epopt(results_path + files[11] + "/train_curve.txt")
epopt2 = read_epopt(results_path + files[12] + "/train_curve.txt")
epopt3 = read_epopt(results_path + files[13] + "/train_curve.txt")
epopt4 = read_epopt(results_path + files[14] + "/train_curve.txt")
epopt = pd.concat([epopt, epopt1, epopt2, epopt3, epopt4])
epopt["time_total_s"] = [round(i, -2) for i in epopt["time_total_s"]]
print(files[10])
print(files[11])
print(files[12])
print(files[13])
print(files[14])
epopt
```

```python
def read_urbe(path):
    df = pd.read_csv(path)
    start_time = df["Wall time"][0]
    df["Wall time"] = [i - start_time for i in df["Wall time"]]
    df["Value"] = [i if i == 200 else (i-500) for i in df["Value"]]
    return df
urbe = read_urbe(results_path + files[20])
urbe1 = read_urbe(results_path + files[21])
urbe2 = read_urbe(results_path + files[22])
urbe3 = read_urbe(results_path + files[23])
urbe4 = read_urbe(results_path + files[24])
urbe = pd.concat([urbe, urbe1, urbe2, urbe3, urbe4])
urbe["Wall time"] = [round(i / 500.0) * 500.0 for i in urbe["Wall time"]]
print(files[20])
print(files[21])
print(files[22])
print(files[23])
print(files[24])
urbe
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

plt.figure(figsize=(9,3), dpi=100)
# gs = fig.add_gridspec(2, 2, hspace=0, wspace=0)
# ax = gs.subplots(sharex='col', sharey='row')
palette=sns.color_palette()
sns.lineplot(data=robust, x="time_total_s", y="episode_reward_mean", color=palette[3], label="Robust DQN", ci=95)
sns.lineplot(data=urbe, x="Wall time", y="Value", color=palette[4], label="DQN-URBE", ci=95)
# sns.lineplot(data=epopt, x="time_total_s", y="10p", color=palette[6], label="EPOpt_10p", ci=95)
sns.lineplot(data=epopt, x="time_total_s", y="average", color=palette[5], label="EPOpt", ci=95)
sns.lineplot(data=deterministic, x="time_total_s", y="episode_reward_mean", color=palette[6], label="Deterministic DQN", ci=95)
plt.xlabel('Time (s)')
plt.ylabel('Reward received in\neach episode (dimensionless)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,3), dpi=100)
palette=sns.color_palette()
# ea = ea[ea["time_total_s"]<=17500]
sns.lineplot(data=ea, x="time_total_s", y="epsilon_progress", color=palette[1], label="Epsilon-NSGA-II", ci=95)
plt.xlabel('Time (s)')
plt.ylabel('Epsilon progress (dimensionless)')
plt.legend()
plt.show()
```

```python
def time_average_DQN(path):
    value = pd.read_csv(path)[["episodes_total", "time_total_s"]].tail(1)
    value = value["time_total_s"] / value["episodes_total"]
    return value.values[0]
def time_average_URBE(path):
    df = pd.read_csv(path)
    start_time = df["Wall time"][0]
    df["Wall time"] = [i - start_time for i in df["Wall time"]]
    return df["Wall time"].tail(1).values[0] / (4000*100)
def time_average_EPOpt(path):
    lines = []
    with open(path, "r") as f:
        for each_line in f:
            line = each_line.strip().split(" ")
            output_line = [float(line[0]), float(line[3]), float(line[5])]
            lines.append(output_line)
    df = pd.DataFrame(lines, columns=["average","10p", "time_total_s"])
    return df["time_total_s"].tail(1).values[0] / (4000*100)
ea_per = np.mean([0.192, 0.1992, 0.2076, 0.1842, 0.1968]) / 100
deterministic_per = np.mean([time_average_DQN(results_path + files[0] + "/progress.csv"), 
                            time_average_DQN(results_path + files[1] + "/progress.csv"),
                            time_average_DQN(results_path + files[2] + "/progress.csv"),
                            time_average_DQN(results_path + files[3] + "/progress.csv"),
                            time_average_DQN(results_path + files[4] + "/progress.csv")])
robust_per = np.mean([time_average_DQN(results_path + files[15] + "/progress.csv"), 
                      time_average_DQN(results_path + files[16] + "/progress.csv"),
                      time_average_DQN(results_path + files[17] + "/progress.csv"),
                      time_average_DQN(results_path + files[18] + "/progress.csv"),
                      time_average_DQN(results_path + files[19] + "/progress.csv")])
urbe_per = np.mean([time_average_URBE(results_path + files[20]),
                    time_average_URBE(results_path + files[21]),
                    time_average_URBE(results_path + files[22]),
                    time_average_URBE(results_path + files[23]),
                    time_average_URBE(results_path + files[24])])
epopt_per = np.mean([time_average_EPOpt(results_path + files[10] + "/train_curve.txt"),
                     time_average_EPOpt(results_path + files[11] + "/train_curve.txt"),
                     time_average_EPOpt(results_path + files[12] + "/train_curve.txt"),
                     time_average_EPOpt(results_path + files[13] + "/train_curve.txt"),
                     time_average_EPOpt(results_path + files[14] + "/train_curve.txt")])
time_average_df = pd.DataFrame({"Value":[ea_per, robust_per, urbe_per, epopt_per, deterministic_per],
                                "Algorithms":["Epsilon-NSGA-II", "Robust DQN", "DQN-URBE", 
                                "EPOpt", "Deterministic DQN"]})
time_average_df
```

```python
fig = plt.figure(figsize=(12,4), dpi=1000)
plt.plot(time_average_df["Algorithms"], time_average_df["Value"])
for method, value in zip(time_average_df["Algorithms"], time_average_df["Value"]): 
    plt.text(method, value, str(round(value,3)))
plt.show()
```

```python
def read_urbe(path, files, file_id):
    df = pd.read_csv(path + files[file_id+1])
    length_df = pd.read_csv(path + files[file_id])
    start_time = df["Wall time"][0]
    df["Wall time"] = [i - start_time for i in df["Wall time"]]
    df["Value"] = [df["Value"][i] if length_df["Value"][i] == 200 else (df["Value"][i]-500) for i in range(len(length_df["Value"]))]
    return df
urbe = read_urbe(results_path, files, 20)
urbe1 = read_urbe(results_path, files, 22)
urbe2 = read_urbe(results_path, files, 24)
urbe3 = read_urbe(results_path, files, 26)
urbe4 = read_urbe(results_path, files, 28)
urbe = pd.concat([urbe, urbe1, urbe2, urbe3, urbe4])
urbe["Wall time"] = [round(i / 500.0) * 500.0 for i in urbe["Wall time"]]
print(files[20])
print(files[22])
print(files[24])
print(files[26])
print(files[28])
urbe
```

```python
def read_epopt(path):
    lines = []
    with open(path, "r") as f:
        for each_line in f:
            line = each_line.strip().split(" ")
            output_line = [float(line[0]), float(line[3]), float(line[5])]
            lines.append(output_line)
    df = pd.DataFrame(lines, columns=["average","10p", "time_total_s"])
    return df
epopt = read_epopt(results_path + files[4] + "/train_curve.txt")
epopt["time_total_s"] = [round(i, -2) for i in epopt["time_total_s"]]
print(files[4])
epopt
```

```python
def read_urbe(path, files, file_id):
    df = pd.read_csv(path + files[file_id])
    length_df = pd.read_csv(path + files[file_id])
    start_time = df["Wall time"][0]
    df["Wall time"] = [i - start_time for i in df["Wall time"]]
    return df
urbe = read_urbe(results_path, files, 5)
urbe["Wall time"] = [round(i / 500.0) * 500.0 for i in urbe["Wall time"]]
print(files[5])
urbe
```
