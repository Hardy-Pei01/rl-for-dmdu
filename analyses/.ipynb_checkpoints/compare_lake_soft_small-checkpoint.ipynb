# Multi-objective Lake Problem

```python
import os, sys
sys.path.append(os.path.abspath('../'))
import pandas as pd
import numpy as np
from EMAWorkbench import load_results

results_path='../results/lake_soft_performance_small/'
files=os.listdir(results_path)
files
```

```python
def compute_error_bar(df):
    u_mean=np.mean(all_outcomes["utility"])
    u_25=np.percentile(all_outcomes["utility"], 25)
    u_75=np.percentile(all_outcomes["utility"], 75)
    r_mean=np.mean(all_outcomes["reliability"])
    r_25=np.percentile(all_outcomes["reliability"], 25)
    r_75=np.percentile(all_outcomes["reliability"], 75)
    return (u_mean, (u_mean-u_25, u_75-u_mean), r_mean, (r_mean-r_25, r_75-r_mean))
```

```python
random=[]
for i in range(169, 170):
    experiments, all_outcomes=load_results(results_path + files[i])
    all_outcomes=dict((k, all_outcomes[k]) for k in ('utility', 'reliability'))
    mean_outcome=compute_error_bar(all_outcomes)
    random.append(mean_outcome)
    print(files[i])
random=pd.DataFrame(random)
random.columns=["utility", "u_inter", "reliability", "r_inter"]
random["Algorithms"]="Random"
random
```

```python
ea=[]
for i in range(75, 138):
    experiments, all_outcomes=load_results(results_path + files[i])
    all_outcomes=dict((k, all_outcomes[k]) for k in ('utility', 'reliability'))
    mean_outcome=compute_error_bar(all_outcomes)
    ea.append(mean_outcome)
    print(files[i])
ea=pd.DataFrame(ea)
ea.columns=["utility", "u_inter", "reliability", "r_inter"]
ea["Algorithms"]="Epsilon-NSGA-II"
ea
```

```python
borg=[]
for i in range(0, 65):
    experiments, all_outcomes=load_results(results_path + files[i])
    all_outcomes=dict((k, all_outcomes[k]) for k in ('utility', 'reliability'))
    mean_outcome=compute_error_bar(all_outcomes)
    borg.append(mean_outcome)
    print(files[i])
borg=pd.DataFrame(borg)
borg.columns=["utility", "u_inter", "reliability", "r_inter"]
borg["Algorithms"]="Borg"
borg
```

```python
urbe=[]
for i in range(170, 180):
    experiments, all_outcomes=load_results(results_path + files[i])
    all_outcomes=dict((k, all_outcomes[k]) for k in ('utility', 'reliability'))
    mean_outcome=compute_error_bar(all_outcomes)
    urbe.append(mean_outcome)
    print(files[i])
urbe=pd.DataFrame(urbe)
urbe.columns=["utility", "u_inter", "reliability", "r_inter"]
urbe["Algorithms"]="DQN-URBE"
urbe
```

```python
no=[]
for i in range(168, 169):
    experiments, all_outcomes=load_results(results_path + files[i])
    all_outcomes=dict((k, all_outcomes[k]) for k in ('utility', 'reliability'))
    mean_outcome=compute_error_bar(all_outcomes)
    no.append(mean_outcome)
    print(files[i])
no=pd.DataFrame(no)
no.columns=["utility", "u_inter", "reliability", "r_inter"]
no["Algorithms"]="No Pollution"
no
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.concat([random, ea, borg, free, urbe, epopt, deterministic, no])
plt.figure(figsize=(12,8), dpi= 100)
sns.scatterplot(data=data, x="utility", y="reliability", hue="Algorithms", style="Algorithms", s=100)
plt.xlabel("utility (unit)")
plt.ylabel("reliability (%)")
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(16,16), dpi=100)
gs = fig.add_gridspec(3, 2, hspace=0, wspace=0)
ax = gs.subplots(sharex='col', sharey='row')
_, _, bars = ax[0,0].errorbar(x=ea['utility'], y=ea['reliability'], xerr=zip(*ea['u_inter']), 
                              yerr=zip(*ea['r_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[0,0].errorbar(x=random['utility'], y=random['reliability'], xerr=zip(*random['u_inter']), 
                               yerr=zip(*random['r_inter']), fmt='o', ls='none', label ='Random')
_, _, bars2 = ax[0,0].errorbar(x=no['utility'], y=no['reliability'], xerr=zip(*no['u_inter']), 
                               yerr=zip(*no['r_inter']), fmt='o', ls='none', label ='No Pollution')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]
[bar.set_alpha(0.5) for bar in bars2]

_, _, bars = ax[0,1].errorbar(x=ea['utility'], y=ea['reliability'], xerr=zip(*ea['u_inter']), 
                              yerr=zip(*ea['r_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[0,1].errorbar(x=borg['utility'], y=borg['reliability'], xerr=zip(*borg['u_inter']), 
                               yerr=zip(*borg['r_inter']), fmt='o', ls='none', label ='Borg')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

_, _, bars = ax[1,0].errorbar(x=ea['utility'], y=ea['reliability'], xerr=zip(*ea['u_inter']), 
                              yerr=zip(*ea['r_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[1,0].errorbar(x=free['utility'], y=free['reliability'], xerr=zip(*free['u_inter']), 
                               yerr=zip(*free['r_inter']), fmt='o', ls='none', label ='Robust Model-free DQN')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

_, _, bars = ax[1,1].errorbar(x=ea['utility'], y=ea['reliability'], xerr=zip(*ea['u_inter']), 
                              yerr=zip(*ea['r_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[1,1].errorbar(x=urbe['utility'], y=urbe['reliability'], xerr=zip(*urbe['u_inter']), 
                               yerr=zip(*urbe['r_inter']), fmt='o', ls='none', label ='DQN-URBE')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

_, _, bars = ax[2,0].errorbar(x=ea['utility'], y=ea['reliability'], xerr=zip(*ea['u_inter']), 
                              yerr=zip(*ea['r_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[2,0].errorbar(x=epopt['utility'], y=epopt['reliability'], xerr=zip(*epopt['u_inter']), 
                               yerr=zip(*epopt['r_inter']), fmt='o', ls='none', label ='EPOpt')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

_, _, bars = ax[2,1].errorbar(x=ea['utility'], y=ea['reliability'], xerr=zip(*ea['u_inter']), 
                              yerr=zip(*ea['r_inter']), fmt='o', ls='none', label ="Epsilon-NSGA-II")
_, _, bars1 = ax[2,1].errorbar(x=deterministic['utility'], y=deterministic['reliability'], xerr=zip(*deterministic['u_inter']), 
                               yerr=zip(*deterministic['r_inter']), fmt='o', ls='none', label ='Deterministic Model-free DQN')
[bar.set_alpha(0.5) for bar in bars]
[bar.set_alpha(0.5) for bar in bars1]

ax[0,0].set_title("(a)")
ax[0,1].set_title("(b)")
ax[1,0].set_title("(c)")
ax[1,1].set_title("(d)")
ax[2,0].set_title("(e)")
ax[2,1].set_title("(f)")
ax[0,0].set(ylabel='reliability (%)')
ax[1,0].set(ylabel='reliability (%)')
ax[2,0].set(xlabel='utility (unit)', ylabel='reliability (%)')
ax[2,1].set(xlabel='utility (unit)')
ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
ax[1,1].legend()
ax[2,0].legend()
ax[2,1].legend()
plt.show()
```
