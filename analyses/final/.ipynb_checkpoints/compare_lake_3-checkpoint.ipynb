```python
import os, sys
sys.path.append(os.path.abspath('../../'))
import pandas as pd
import numpy as np
from EMAWorkbench import load_results
from utils.rl.hv import HyperVolume
import matplotlib.pyplot as plt


def compute_hypervolume(df, antiutopia, utopia):
    array = df.values
    hv_computer = HyperVolume(antiutopia, utopia)
    return hv_computer.compute(array)
```

# Discrete 4-objective Lake Problem

```python
results_path='../../results/lake4_discrete_performance_2/'
files = [f for f in sorted(os.listdir(results_path)) if not f.startswith('.')]

utopia = np.array([0, 1.8, 1, 1])
antiutopia = np.array([-2.1, 0, 0, 0])

files
```

```python
ea_hv = []
for i in range(0, 20):
    ea=[]
    ea_files = sorted(os.listdir(results_path + files[i]))
    for j in range(len(ea_files)):
        _, all_outcomes=load_results(results_path + files[i] + "/" + ea_files[j])
        all_outcomes=dict((k, np.mean(all_outcomes[k])) for k in ("avg_pollution", "utility", "inertia", "reliability"))
        ea.append(all_outcomes)
        # print(ea_files[j])
    ea=pd.DataFrame(ea)
    ea.iloc[:, 0] = -ea.iloc[:, 0]
    each_ea_hv = compute_hypervolume(ea, antiutopia, utopia)
    ea_hv.append(each_ea_hv)
    print(files[i])
    # print(ea)
    print(each_ea_hv)
ea_hv = np.array(ea_hv)
print(ea_hv.mean())
print(ea_hv.std())
```

```python
import copy
from matplotlib.lines import Line2D

lines = [Line2D([0], [0], color=c, linewidth=2, linestyle='-') for c in ["blue", "red", "green"]]
labels = ["$\epsilon$-NSGA-II", "MO-NES", "MO-eREPS"]

def discrete_lake_diagram(df, lines, labels):
    
    df = copy.deepcopy(df)
    df.iloc[:, 0] = 2.1-df.iloc[:, 0]
    df.iloc[:, 1] = 2.1*df.iloc[:, 1]/1.8
    df.iloc[:, 2] = 2.1*df.iloc[:, 2]
    df.iloc[:, 3] = 2.1*df.iloc[:, 3]

    fig,(ax1, ax2, ax3) = plt.subplots(1, 3, sharey=False)
    ax4 = ax3.twinx()

    x = ['Avg Pollution (dimensionless)', 'Utility (dimensionless)', 'Inertia (dimensionless)', 'Reliability (dimensionless)']

    colors = {"$\epsilon$-NSGA-II":"blue",
              "MO-NES":"red",
              "MO-eREPS":"green"}
    for i in range(df.shape[0]):
        ax1.plot(x, df.iloc[i, :-1], color=colors[df.iloc[i, -1]])
        ax2.plot(x, df.iloc[i, :-1], color=colors[df.iloc[i, -1]])
        ax3.plot(x, df.iloc[i, :-1], color=colors[df.iloc[i, -1]])

    # now zoom in each of the subplots
    ax1.set_xlim([x[0], x[1]])
    ax2.set_xlim([x[1], x[2]])
    ax3.set_xlim([x[2], x[3]])

    ax1.set_ylim([0, 2.25])
    ax2.set_ylim([0, 2.25])
    ax3.set_ylim([0, 2.25])
    ax4.set_ylim([0, 2.25])

    ax1.set_yticks((0, 2.1))
    ax1.set_yticklabels(("2.1", "0"))

    ax2.set_yticks((0, 2.1))
    ax2.set_yticklabels(("0", "1.8"))

    ax3.set_yticks((0, 2.1))
    ax3.set_yticklabels(("0", "1"))

    ax4.set_yticks((0, 2.1))
    ax4.set_yticklabels(("0", "1"))

    # stack the subplots together
    plt.subplots_adjust(wspace=0)
    ax1.legend(lines, labels, loc='upper left')
    plt.show()
```

```python
ax = pd.plotting.parallel_coordinates(
    ea, 'Algorithms', color=('blue')
)
ax.set_ylim(0, 2.1)
```

```python
import copy
from matplotlib.lines import Line2D

lines = [Line2D([0], [0], color=c, linewidth=2, linestyle='-') for c in ["blue", "red", "green"]]
labels = ["$\epsilon$-NSGA-II", "MO-NES", "MO-eREPS"]

def robust_lake_diagram(df, lines, labels):
    
    df = copy.deepcopy(df)
    df.iloc[:, 0] = 4-df.iloc[:, 0]
    df.iloc[:, 1] = 4*df.iloc[:, 1]/1.25
    df.iloc[:, 2] = 4*df.iloc[:, 2]
    df.iloc[:, 3] = 4*df.iloc[:, 3]

    fig,(ax1, ax2, ax3) = plt.subplots(1, 3, sharey=False)
    ax4 = ax3.twinx()

    x = ['Avg Pollution (dimensionless)', 'Utility (dimensionless)', 'Inertia (dimensionless)', 'Reliability (dimensionless)']

    colors = {"$\epsilon$-NSGA-II":"blue",
              "MO-NES":"red",
              "MO-eREPS":"green"}
    for i in range(df.shape[0]):
        ax1.plot(x, df.iloc[i, :-1], color=colors[df.iloc[i, -1]])
        ax2.plot(x, df.iloc[i, :-1], color=colors[df.iloc[i, -1]])
        ax3.plot(x, df.iloc[i, :-1], color=colors[df.iloc[i, -1]])

    # now zoom in each of the subplots
    ax1.set_xlim([x[0], x[1]])
    ax2.set_xlim([x[1], x[2]])
    ax3.set_xlim([x[2], x[3]])

    ax1.set_ylim([0, 4.25])
    ax2.set_ylim([0, 4.25])
    ax3.set_ylim([0, 4.25])
    ax4.set_ylim([0, 4.25])

    ax1.set_yticks((0, 4))
    ax1.set_yticklabels(("4", "0"))

    ax2.set_yticks((0, 4))
    ax2.set_yticklabels(("0", "1.25"))

    ax3.set_yticks((0, 4))
    ax3.set_yticklabels(("0", "1"))

    ax4.set_yticks((0, 4))
    ax4.set_yticklabels(("0", "1"))

    # stack the subplots together
    plt.subplots_adjust(wspace=0)
    ax1.legend(lines, labels, loc='upper left')
    plt.show()
```
