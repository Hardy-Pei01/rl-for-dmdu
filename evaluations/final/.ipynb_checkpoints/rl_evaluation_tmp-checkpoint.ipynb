# Discrete 4-objective Lake Problem

```python
import os, sys
sys.path.append(os.path.abspath('../../'))
import re, pickle
import numpy as np
import pandas as pd
from algs.RUN_Manifold import run, eval
from models.MOMDP import MOMDP
from library.basis.lake4_basis_poly import lake4_basis_poly
from library.policies.Gibbs import Gibbs

policy_path = '../../results/rl_lake_42/'
files = [f for f in sorted(os.listdir(policy_path)) if not f.startswith('.')]
files
```

```python
for j in range(2):
    
    for name in ["best", "high"]:
        hv_lis = []
        
        for i in range(j*40, j*40+40, 2):
            tmp = re.split("_|-", files[i])
            label = f'{tmp[0]}'

            #  ======================== LOW LEVEL SETTINGS =========================  #
            random_seed = 1793476144
            problem = 'lake4_discrete'
            r_metric = "avg"
            utopia = np.array([0, 1.8, 1, 1])
            antiutopia = np.array([-2.1, 0, 0, 0])
            mdp = MOMDP(problem, seed=random_seed, utopia=utopia, antiutopia=antiutopia, dstate=4, dreward=4)
            robj = 1

            bfs = lake4_basis_poly

            policy = Gibbs(bfs, np.zeros(((bfs() + 1) * mdp.actionUB, 1)), np.arange(mdp.actionLB, mdp.actionUB + 1))
            policy.makeDeterministic()
            #  ======================== LEARNING SETTINGS ==========================  #
            episodes_eval = 1
            steps_eval = 100

            final_policy_path = f'{policy_path}{files[i][:-12]}'
            print(f'{final_policy_path}_{name}.pickle')
            with open(f'{final_policy_path}_{name}.pickle', 'rb') as fh:
                best_policy = pickle.load(fh)

            N_EVAL = 2000
            hv, f = eval(N_EVAL, mdp, episodes_eval, steps_eval, policy, best_policy, None, None, None, r_metric=r_metric)
            hv_lis.append(hv)

        results = pd.DataFrame(hv_lis).T
        result_path=f'../../results/lake4_discrete_performance_2/'
        results.to_csv(result_path + f'{label}_{name}.csv', index=False)
        print(result_path + f'{label}_{name}.csv')
```

# Robust 4-objective Lake Problem

```python
import os, sys
sys.path.append(os.path.abspath('../../'))
import re, pickle
import numpy as np
import pandas as pd
from algs.RUN_Manifold import run, eval
from models.MOMDP import MOMDP
from library.basis.lake4_basis_poly import lake4_basis_poly
from library.policies.Gibbs import Gibbs

policy_path = '../../results/rl_lake_robust_43/'
files = [f for f in sorted(os.listdir(policy_path)) if not f.startswith('.')]
files
```

```python
for j in range(2):
    
    for name in ["best", "high"]:
        hv_lis = []

        for i in range(j*2, j*2+2, 2):
            tmp = re.split("_|-", files[i])
            label = f'{tmp[0]}_{tmp[1]}'
            
            #  ======================== LOW LEVEL SETTINGS =========================  #
            random_seed = 1793476144
            problem = 'lake4_deep_uncertain'
            r_metric = "avg"
            utopia = np.array([0, 2.6, 1, 1])
            antiutopia = np.array([-10, 0, 0, 0])
            mdp = MOMDP(problem, seed=random_seed, utopia=utopia, antiutopia=antiutopia, dstate=4, dreward=4)
            robj = 1

            bfs = lake4_basis_poly

            policy = Gibbs(bfs, np.zeros(((bfs() + 1) * mdp.actionUB, 1)), np.arange(mdp.actionLB, mdp.actionUB + 1))
            policy.makeDeterministic()
            #  ======================== LEARNING SETTINGS ==========================  #
            episodes_eval = 1000
            steps_eval = 100

            final_policy_path = f'{policy_path}{files[i][:-12]}'
            print(f'{final_policy_path}_{name}.pickle')
            with open(f'{final_policy_path}_{name}.pickle', 'rb') as fh:
                best_policy = pickle.load(fh)

            N_EVAL = 2000
            evaluation_scenarios = pd.read_csv("../../results/lake_scenarios/evaluation_scenarios.csv")
            hv, f = eval(N_EVAL, mdp, episodes_eval, steps_eval, policy, best_policy, 
                         None, None, evaluation_scenarios, r_metric=r_metric)
            hv_lis.append(hv)

        results = pd.DataFrame(hv_lis).T
        result_path=f'../../results/lake4_robust_performance_1/'
        results.to_csv(result_path + f'{label}_{name}.csv', index=False)
        print(result_path + f'{label}_{name}.csv')
```
